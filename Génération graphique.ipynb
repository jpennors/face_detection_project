```python
import os 
dir = "comparaison_modeles"
# Random Forest

x = [25,50,75,100,125,150, 200, 250]
y = [0.785,0.810,0.847,0.836,0.839,0.856, 0.88, 0.85]

import matplotlib.pyplot as plt
plt.plot(x,y)
plt.title("Evolution de la précision moyenne en fonction du paramètre\n \"n_estimator\" pour le modèle Forêt aléatoire")
plt.xlabel("n_estimator")
plt.ylabel("Précision moyenne")

if not os.path.exists(dir):
        os.makedirs(dir)
    
plt.savefig(dir + "/random-forest.png")
```
